### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

````python
import pandas as pd
order_products_with_names = order_products.merge(products[['product_id', 'product_name']], on='product_id', how='left')
product_reorder = order_products_with_names.groupby(['product_id', 'product_name']).agg(
    total_orders = ('reordered', 'count'),
    total_reorders = ('reordered', 'sum')
).reset_index()
product_reorder['repeat_rate'] = product_reorder['total_reorders'] / product_reorder['total_orders']
top_repeat_products = product_reorder.sort_values(by='repeat_rate', ascending=False).head(20)
print(top_repeat_products)
````
